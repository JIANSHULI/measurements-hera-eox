In this notebook, I check whether the presence of washers connecting the two halves of the balun and through board have an impact on the measurement. 

To do this, I measure $S_{12}$ and $S_{22}$ of the balun using the connections in the through-board with a 50$\Omega$ termination at the un-balanced port.

![title](balun_measurement_no_washers.jpg)

```python
rootDir='../washer_test_July_12th_2017/'
s12_pha_nw=gD.readAnritsuCSV(rootDir+'s12_nw_pha.csv')
s12_amp_nw=gD.readAnritsuCSV(rootDir+'s12_nw_amp.csv')
s12_pha_w=gD.readAnritsuCSV(rootDir+'s12_w_pha.csv')
s12_amp_w=gD.readAnritsuCSV(rootDir+'s12_w_amp.csv')
s12_nw=10**(-s12_amp_nw[:,1]/10.)*np.exp(1j*np.radians(s12_pha_nw[:,1]))
s12_w=10**(-s12_amp_w[:,1]/10.)*np.exp(1j*np.radians(s12_pha_w[:,1]))
```

```python
plt.plot(s12_pha_w[:,0],10.*np.log10(np.abs(s12_w)),s12_pha_nw[:,0],10.*np.log10(np.abs(s12_nw)))
```

```python
plt.plot(s12_pha_w[:,0],np.unwrap(np.angle(s12_w)),s12_pha_nw[:,0],np.unwrap(np.angle(s12_nw)))
```
